### Step 4: Loading Batches With Dataloader

```python
torch.manual_seed(100)

train_loader = DataLoader(train_data, batch_size=100, shuffle=True)

test_loader = DataLoader(test_data, batch_size=500, shuffle=False)
```

### Step 6: Defining the Neural Network Class

```python
class NeuralNetwork(nn.Module):
    def __init__(self, in_size=28*28, out_size=10, layers=[128,64]):
        super().__init__()
        self.linear_relu_stack = nn.Sequential( # Defining the layers and the activation function
            nn.Linear(in_size, layers[0]),
            nn.ReLU(),
            nn.Linear(layers[0], layers[1]),
            nn.ReLU(),
            nn.Linear(layers[1], out_size)   
        )
        
    def forward(self, X): # Forward propagation 
        logits = self.linear_relu_stack(X)
        return F.log_softmax(logits, dim=1)
```

```python
torch.manual_seed(100)
model = NeuralNetwork()
model
```

### Step 7: Count the Model Parameters

```python
def count_parameters(model):
    params = [p.numel() for p in model.parameters() if p.requires_grad]
    for item in params:
        print(f'{item:>6}')
    print(f'______\n{sum(params):>6}')
```

```python
count_parameters(model)
```

### Step 8: Defining the Loss Function and the Optimizer

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

### Step 9: Training the Model

```python
import time
```

```python
start_time = time.time()

epochs = 10
train_losses = []
train_correct = []
test_losses = []
test_correct = []

for i in range(epochs):
    trn_corr = 0 
    tst_corr = 0

    # Running the training batches
    for b, (X_train, y_train) in enumerate(train_loader):
        b+=1
        # Applying the model 
        y_pred = model(flatten(X_train))
        loss = criterion(y_pred, y_train)
        # Tallying the number of correct predictions
        predicted = torch.max(y_pred.detach(), 1)[1]
        batch_corr = (predicted == y_train).sum()
        trn_corr += batch_corr

        # Updating parameteres
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Print interim results
        if b%200 == 0:
            print(f'epoch: {i:2}  batch: {b:4} [{100*b:6}/60000]  loss: {loss.item():10.8f}  \
            accuracy: {trn_corr.item()*100/(100*b):7.3f}%')

        # Updating train loss and accuracy for each epoch
    train_losses.append(loss.item())
    train_correct.append(trn_corr.item())

        # Running the testing batches 
    with torch.no_grad():
        for b, (X_test, y_test) in enumerate(test_loader):
            y_val = model(flatten(X_test))
            # Tallying the number of correct predictions 
            predicted = torch.max(y_val.detach(),1)[1]
            tst_corr += (predicted == y_test).sum()
                
    # Update the test loss and accuracy for each epoch
    loss = criterion(y_val, y_test)
    test_losses.append(loss.item())
    test_correct.append(tst_corr.item())

print(f'\nDuration: {time.time()-start_time:.0f} seconds')    
```

### Step 10: Plotting the Loss and Accuracy Comparions

```python
plt.plot(train_losses, label='training loss')
plt.plot(test_losses, label='validation loss')
plt.title('Loss at the end of each epoch')
plt.legend()
```

```python
plt.plot([t/600 for t in train_correct], label='training accuracy')
plt.plot([t/100 for t in test_correct], label='validation accuracy')
plt.title('Accuracy at the end of each epoch')
plt.legend();
```
